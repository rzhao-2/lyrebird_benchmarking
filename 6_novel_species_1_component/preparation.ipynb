```python
import polars as pl
from os.path import join
from os import listdir
```

```python
lyrebird_dir = "output_lyrebird"
metaphlan_dir = "output_metaphlan4"
phanta_dir = "output_phanta"
```

```python
genomes = []
with open("genome_rep_tax.csv") as f:
    for line in f:
        genomes.append(line.strip().split('\t')[0])
```

```python
lyrebird_count = 0
metaphlan_count = 0
phanta_count = 0
for genome in genomes:
    if f"{genome}.profile" in listdir(join(lyrebird_dir)):
        lyrebird_count += 1
    if f"{genome}.profile" in listdir(join(metaphlan_dir)):
        with open(join(metaphlan_dir, f"{genome}.profile")) as f:
            if f.readline().startswith("#"):
                metaphlan_count += 1
    if f"{genome}_output_phanta" in listdir(join(phanta_dir)):
        if "final_merged_outputs" in listdir(join(phanta_dir, f"{genome}_output_phanta")):
            if "relative_taxonomic_abundance.txt" in listdir(join(phanta_dir, f"{genome}_output_phanta", "final_merged_outputs")):
                with open(join(phanta_dir, f"{genome}_output_phanta", "final_merged_outputs", "relative_taxonomic_abundance.txt")) as f:
                    if len(f.readlines()) > 1:
                        phanta_count += 1

print(f"Lyrebird: {lyrebird_count}")
print(f"Metaphlan4: {metaphlan_count}")
print(f"Phanta: {phanta_count}")
with open("genome_counter.csv", "w") as f:
    f.write("tool,count\n")
    f.write(f"lyrebird,{lyrebird_count}\n")
    f.write(f"metaphlan4,{metaphlan_count}\n")
    f.write(f"phanta,{phanta_count}\n")
    # f.write("lyrebird,metaphlan4\n")
    # f.write(f"{lyrebird_count},{metaphlan_count}\n")
```

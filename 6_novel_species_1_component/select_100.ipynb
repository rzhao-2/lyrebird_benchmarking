```python
import random
reps_list = "/mnt/hpccs01/work/microbiome/msingle/rossenzhao/lyrebird_novelty/galah_ictv_reps.txt"
taxonomy = "/mnt/hpccs01/work/microbiome/msingle/rossenzhao/lyrebird_novelty/lyrebird_novel_tax.txt"
outfile = "genome_rep_paths.csv"
```

```python
genome_to_fp = []
with open(reps_list, 'r') as f:
    for line in f:
        filepath = line.strip()
        genome = filepath.split("/")[-1].split(".")[0]
        genome_to_fp.append(f'{genome}\t{filepath}\n')

random.shuffle(genome_to_fp)
with open(outfile, 'w') as f:
    for i in range(0,100):
        f.write(genome_to_fp[i])
```

```python
outtax = "genome_rep_tax.csv"
genome_to_tax = []
with open(taxonomy, 'r') as f:
    for line in f:
        genome, tax = line.strip().split("\t")
        genome_to_tax.append(f'{genome}\t{tax}\n')

with open(outtax, 'w+') as f:
    with open(outfile, 'r') as g:
        for line in g:
            genome = line.strip().split("/")[-1].split(".")[0]
            for entry in genome_to_tax:
                if genome in entry:
                    f.write(entry)
```

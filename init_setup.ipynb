```python
import os
import extern
# import polars as pl
# import random as rand

roundrobin_species_coverages = "/mnt/hpccs01/work/microbiome/msingle/rossenzhao/phrogs_v4.1/build_spkgs/roundrobin_species_coverage.tsv"
genome_rep_paths = "/home/n10927662/lyrebird-benchmarking/genome_rep_paths.csv"
```

```python
### 1. Sanity check at 10x coverage
sanity_genomes_output = "/home/n10927662/lyrebird-benchmarking/1_sanity_10x/genome_rep_paths.csv"

genome_to_paths = {}
with open(genome_rep_paths) as r:
    for line in r:
        genome, path = line.strip().split("\t")
        genome_to_paths[genome] = path

num_over_20 = []
num_10_20 = []
num_3 = []
with open(sanity_genomes_output, "w+") as w:
    with open(roundrobin_species_coverages, "r") as f:
        for line in f:
            species, coverage = line.strip().split("\t")
            coverage = int(coverage)
            if "IMGVR" in species:
                continue
            if coverage >= 20:
                if len(num_over_20) < 3:
                    num_over_20.append(species)
            elif coverage >= 10:
                if len(num_10_20) < 3:
                    num_10_20.append(species)
            elif coverage > 3:
                if len(num_3) < 3:
                    num_3.append(species)
            if len(num_over_20) >= 3 and len(num_10_20) >= 3 and len(num_3) >= 3:
                break

    for species in num_over_20:
        w.write(f"{species}\t{genome_to_paths[species]}\n")
    for species in num_10_20:
        w.write(f"{species}\t{genome_to_paths[species]}\n")
    for species in num_3:
        w.write(f"{species}\t{genome_to_paths[species]}\n")
```

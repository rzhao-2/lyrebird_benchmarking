```r
library(ggplot2)
library(data.table)
library(reshape2)
```

```r
df = read.csv("lyrebird_for_vln.txt", sep="\t")
```

```r
options(repr.plot.width = 10, repr.plot.height = 10)
df_melt <- melt(df, measure.vars = c("TPR", "Precision", "FPR"), variable.name = "Metric", value.name = "Value")

p <- ggplot(df_melt, aes(x = Metric, y = Value, fill = Metric)) +
    geom_violin(trim = FALSE, alpha = 0.7) +
    scale_fill_manual(values = c("TPR" = "#00BA38", "Precision" = "#619CFF", "FPR" = "#F8766D")) +
    labs(title = "Lyrebird accuracy metrics on simulated phage-microbe pairs", x = "", y = "Value", fill = "Metric") +
    ylim(0, NA) +
    theme_minimal(base_size = 16) +
    theme(legend.position = "none", axis.text.x = element_text(size = 16), axis.text.y = element_text(size = 16))

print(p)
```

```r
p_precision <- ggplot(df_melt[df_melt$Metric == "Precision", ], aes(x = Metric, y = Value, fill = Metric)) +
    geom_violin(trim = FALSE, alpha = 0.7) +
    scale_fill_manual(values = c("Precision" = "#619CFF")) +
    labs(title = "Lyrebird precision on simulated phage-microbe pairs", x = "", y = "Value") +
    ylim(0, NA) +
    theme_minimal(base_size = 16) +
    theme(legend.position = "none", axis.text.x = element_text(size = 16), axis.text.y = element_text(size = 16))

print(p_precision)
```

```python
import os
from tqdm import tqdm
from Bio import SeqIO
provirus_ids = "/mnt/hpccs01/home/n10927662/misc_scripts/create_metapackage/gtdb_provirus_ids_r214.tsv"
microbial_rep_paths = "/home/n10927662/rossenzhao/lyrebird-benchmarking/5_sanity_microbial/microbial_rep_paths.csv"
```

```python
proviruses = set()
with open(provirus_ids) as r:
    for line in r:
        proviruses.add(line.strip())

microbes_without_proviruses = []

with open(microbial_rep_paths) as r:
    for line in tqdm(r):
        id, filepath = line.strip().split("\t")
        for record in SeqIO.parse(filepath, "fasta"):
            if record.id in proviruses:
                break
        else:
            microbes_without_proviruses.append(line)
```

```python
with open("/home/n10927662/rossenzhao/lyrebird-benchmarking/5_sanity_microbial/microbial_rep_paths_no_proviruses.csv", "w+") as w:
    for line in microbes_without_proviruses:
        w.write(line)
```

```python
import os
from os.path import join
### generate viral rep paths
roundrobin_covs = "/mnt/hpccs01/work/microbiome/msingle/rossenzhao/phrogs_v4.1/0.3/roundrobin_species_coverage.tsv"
genome_dir = "/mnt/hpccs01/work/microbiome/msingle/rossenzhao/ictv-imgvr/genomes/"

with open(roundrobin_covs) as r:
    with open("viral_rep_paths.tsv", "w+") as w:
        for line in r:
            genome_id, cov = line.strip().split("\t")
            if genome_id.startswith("NC_"):
                if int(cov) >= 10:
                    filepath = join(genome_dir, genome_id + ".fna")
                    if os.path.exists(filepath):
                        w.write(f"{genome_id}\t{filepath}\n")
                    else:
                        print(f"File not found: {filepath}")
```
